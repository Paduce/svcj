```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, timedelta, time
from ics import Calendar

# ─── CONFIGURATION ─────────────────────────────────────────────────────────────
ICS_FILE = "/home/pduce/Downloads/Takeout/Calendar/pduce@swissblock.net.ics"                   # Path to your .ics file
TARGET_EMAIL = "skorzen@swissblock.net"         # Replace with the email to filter
WORK_START = time(9, 0)
WORK_END = time(18, 0)

# ─── STEP 1: LOAD AND FILTER MEETINGS ──────────────────────────────────────────

# Load ICS file
with open(ICS_FILE, "r", encoding="utf-8") as f:
    cal = Calendar(f.read())

# Parse events that contain the target email
events = []
for event in cal.events:
    attendees = getattr(event, "attendees", [])
    if any(TARGET_EMAIL.lower() in str(a).lower() for a in attendees):
        events.append({
            "summary": event.name,
            "start": event.begin.datetime.astimezone(tz=None).replace(tzinfo=None),
            "end": event.end.datetime.astimezone(tz=None).replace(tzinfo=None),
        })

# Sort by meeting end time
df = pd.DataFrame(events).sort_values(by="end").reset_index(drop=True)
df = df[df['summary'].isin(['Pablo  / Simon [update]','Pablo  / Simon', 'Simon / Pablo '])]
df = df[df['start'] >='2024-10-01'].reset_index(drop=True)
# ─── STEP 2: WORKING-TIME GAP CALCULATION ──────────────────────────────────────

def is_weekday(dt):
    return dt.weekday() < 5  # Monday–Friday

def working_interval(dt):
    return (
        datetime.combine(dt.date(), WORK_START),
        datetime.combine(dt.date(), WORK_END)
    )

def working_time_diff(start: datetime, end: datetime) -> timedelta:
    """Compute working hours (9–18, Mon–Fri) between two datetimes, excluding weekends."""
    if start >= end:
        return timedelta(0)

    total = timedelta(0)
    current = start

    while current.date() <= end.date():
        if is_weekday(current):
            work_start, work_end = working_interval(current)

            # Clip interval to valid working hours
            interval_start = max(current, work_start)
            interval_end = min(end, work_end)

            if interval_start < interval_end:
                total += interval_end - interval_start

        # Move to next day at 9:00
        current = datetime.combine(current.date() + timedelta(days=1), WORK_START)

    return total

def working_days_diff(start: datetime, end: datetime) -> int:
    """Return number of working days (Mon–Fri) between two datetimes."""
    if start >= end:
        return 0

    start_date = start.date()
    end_date = end.date()

    delta = timedelta(days=1)
    current = start_date + delta  # do not count the day of the first meeting
    count = 0

    while current < end_date:
        if current.weekday() < 5:
            count += 1
        current += delta

    return count

# ─── STEP 3: COMPUTE WORKING TIME GAPS ─────────────────────────────────────────

gaps = [None]
for i in range(1, len(df)):
    end_prev = df.loc[i - 1, "end"]
    start_curr = df.loc[i, "start"]
    gap_days = working_days_diff(end_prev, start_curr)
    gaps.append(gap_days)
df["working_days_gap"] = gaps

# ─── STEP 4: PLOT GAPS AND MEETING COUNTS ──────────────────────────────────────


# Plot: Working days between meetings
plt.figure(figsize=(10, 4))
plt.plot(df["start"][1:], df["working_days_gap"][1:], marker="o")
plt.title(f"Working-Day Gaps Between Meetings with {TARGET_EMAIL}")
plt.ylabel("Gap (working days)")
plt.xlabel("Meeting Start Time")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Meetings per week
df["week"] = df["start"].dt.to_period("W").apply(lambda r: r.start_time)
weekly_counts = df.groupby("week").size()

plt.figure(figsize=(10, 4))
weekly_counts.plot(kind="bar")
plt.title(f"Number of Meetings per Week with {TARGET_EMAIL}")
plt.ylabel("Number of Meetings")
plt.xlabel("Week Start")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# ─── OUTPUT SUMMARY ────────────────────────────────────────────────────────────
print(df[["start", "end", "summary", "working_time_gap_hours"]])
```
